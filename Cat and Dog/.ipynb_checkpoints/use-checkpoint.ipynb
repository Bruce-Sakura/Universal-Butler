```python
import os
import numpy as np
from PIL import Image
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
data = "G:\\Photo\\.Cat\\H10A5072.JPG"
```

```python
# 预处理图像函数
def preprocess_image(img_path, new_size=(128, 128)):
    img = Image.open(img_path)  # 打开图像
    img = img.convert('L')  # 转换为灰度图
    img = img.resize(new_size)  # 调整大小
    img_array = np.array(img) / 255.0  # 将图像归一化到[0, 1]范围
    img_array = np.expand_dims(img_array, axis=-1)  # 添加通道维度
    return img_array

# 检查文件是否存在并进行预测
if os.path.isfile(data):
    preprocessed_image = preprocess_image(data)  # 处理图像
    preprocessed_image = np.expand_dims(preprocessed_image, axis=0)  # 添加批次维度

    # 打印输入形状
    print("预处理后的图像形状:", preprocessed_image.shape)  # 应该是 (1, 128, 128, 1)

    # 进行预测
    predictions = loaded_model.predict(preprocessed_image)  # 直接使用预处理后的图像
    
    # 将预测结果转换为标签
    predicted_label = int(predictions[0][0] < 0.5)  # 使用 0.5 作为阈值
    pred_label_text = 'dog' if predicted_label == 0 else 'cat'  # 预测标签

    # 显示图像和预测结果
    plt.figure(figsize=(5, 5))
    plt.imshow(preprocessed_image[0].reshape(128, 128), cmap='gray')  # 去掉批次维度
    plt.title(f'Predicted: {pred_label_text}')
    plt.axis('off')
    plt.show()
else:
    print("文件不存在，请检查路径。")
```
