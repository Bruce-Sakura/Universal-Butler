```python
import torch
import torch.nn as nn

# 假设图像特征是一个大小为 (batch_size, img_feature_dim) 的张量
# 假设文本特征是一个大小为 (batch_size, text_feature_dim) 的张量
batch_size = 4
img_feature_dim = 128  # 图像特征维度
text_feature_dim = 64   # 文本特征维度

# 随机生成图像特征和文本特征
image_features = torch.randn(batch_size, img_feature_dim)  # 生成随机图像特征
text_features = torch.randn(batch_size, text_feature_dim)   # 生成随机文本特征

# 拼接特征
combined_features = torch.cat((image_features, text_features), dim=1)  # 在特征维度上拼接

# 打印拼接后的特征维度
print("Combined feature shape:", combined_features.shape)  # 输出应为 (batch_size, img_feature_dim + text_feature_dim)
```

```python
import torch
import torch.nn as nn

# 假设输入特征维度为 192，输出维度为 10（例如 10 个类别）
input_dim = 192
output_dim = 10

# 定义全连接层
fc_layer = nn.Linear(input_dim, output_dim)

# 创建一个随机输入
input_features = torch.randn(4, input_dim)  # 4 个样本

# 通过全连接层处理输入
output = fc_layer(input_features)

print("Output shape:", output.shape)  # 应输出 (4, 10)
```
